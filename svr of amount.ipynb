```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVR
```

```python
df = pd.read_csv("TaxiFare.csv")
```

```python
# Generate sample data
X = np.sort(5 * np.random.rand(40, 1), axis=0)
y = np.sin(X).ravel()
```

```python
# Add some noise to the output
y[::5] += 3 * (0.5 - np.random.rand(8))
```

```python
# Fit SVR models with different kernels
svr_lin = SVR(kernel='linear', C=100, gamma='auto')
svr_poly = SVR(kernel='poly', C=100, gamma='auto', degree=3, epsilon=0.1, coef0=1)
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_lin.fit(X, y)
svr_poly.fit(X, y)
svr_rbf.fit(X, y)
```

```python
# Predict new values
X_pred = np.linspace(0, 5, 100).reshape(-1, 1)
y_lin = svr_lin.predict(X_pred)
y_poly = svr_poly.predict(X_pred)
y_rbf = svr_rbf.predict(X_pred)
```

```python
# Plot the results
plt.scatter(X, y, color='red', label='data')
plt.plot(X_pred, y_lin, color='navy', label='linear model')
plt.plot(X_pred, y_poly, color='blue', label='polynomial model')
plt.plot(X_pred, y_rbf, color='c', label='RBF model')
plt.xlabel('data')
plt.ylabel('range')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
